```python
#9
r=0.2
xprime=r*x*(1-(x/7)) - (x/(1+x))*y
yprime= (1/10)*y*(1-(y/x))
jacobian1=jacobian([xprime,yprime], [x,y])
show(jacobian1)
```

```python
#10 Substitute the equilibrium values of x and y into the Jacobian you found previously. Then, compute the eigenvalues of the resulting matrix. Use Table 1 to classify this equilibrium.
results=jacobian1.subs ({x:0.239, y:0.24})
results.eigenvectors_right()
show (results) #this is a stable spiral
```
